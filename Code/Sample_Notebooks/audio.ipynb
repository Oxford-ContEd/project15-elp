```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt= 0.001
t=np.arange(0,10,dt)
n=len(t)
f_clean = np.sin(2*np.pi*50*t) + np.sin(2*np.pi*150*t)
f= f_clean+2.5*np.random.randn(n) # random noise added
```

```python
plt.plot(f)
```

```python
fhat = np.fft.fft(f,n)
PSD = (fhat* np.conj(fhat)) /n
```

```python
freq = (1/(dt*n)) * np.arange(n)
L = np.arange(1,np.floor(n/2),dtype="int")
plt.plot(freq[L],PSD[L])
```
